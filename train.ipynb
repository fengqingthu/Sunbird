## Data Preprocessing

```python
import pandas as pd

FILEPATH = '/content/drive/MyDrive/SCI6484/Project/Data/SC_simulation_results/NM_Albuquerque/'
df = pd.read_csv(FILEPATH + "dataset.csv")
print("total number of rows:", len(df.index))
df.head()
```

```python
def remove_outliers(df, column):
    q1 = df[column].quantile(0.25)
    q3 = df[column].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    df = df[(df[column] > lower_bound) & (df[column] < upper_bound)]
    return df

# Remove outliers from each column
columns = df.columns
for column in columns:
    df = remove_outliers(df, column)

print("total number of rows:", len(df.index))
df.head()
```

#Train

### Prepare normalized dataset

```python
import pandas as pd
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.model_selection import train_test_split
from torch.utils.data import DataLoader, Dataset
from sklearn.preprocessing import MinMaxScaler

# Split the data into training and testing sets
train_df, test_df = train_test_split(df, test_size=0.2, random_state=42)

class CustomDataset(Dataset):
    def __init__(self, data):
        self.data = data.values.astype('float32')
        self.x = self.data[:, :5]
        self.y = self.data[:, 5:]

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        x = torch.from_numpy(self.x[idx])
        y = torch.from_numpy(self.y[idx])
        return x, y

# Define the scaler
scaler = MinMaxScaler(feature_range=(0,100))

# Normalize the input data
train_df.iloc[:, :5] = scaler.fit_transform(train_df.iloc[:, :5])
test_df.iloc[:, :5] = scaler.transform(test_df.iloc[:, :5])

# Normalize the output data
train_df.iloc[:, 5:] = scaler.fit_transform(train_df.iloc[:, 5:])
test_df.iloc[:, 5:] = scaler.transform(test_df.iloc[:, 5:])

# Create the training and testing datasets
train_dataset = CustomDataset(train_df)
test_dataset = CustomDataset(test_df)

# Create data loaders for batch processing
batch_size = 8
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)

train_df.head()
```

```python
import torch.nn.functional as F

class Net(nn.Module):
  def __init__(self):
      super(Net, self).__init__()
      self.fc1 = nn.Linear(5, 128)
      self.fc2 = nn.Linear(128, 256)
      self.fc3 = nn.Linear(256, 512)
      self.fc4 = nn.Linear(512, 1024)
      self.fc5 = nn.Linear(1024, 512)
      self.fc6 = nn.Linear(512, 256)
      self.fc7 = nn.Linear(256, 128)
      self.fc8 = nn.Linear(128, 9)

  def forward(self, x):
      x = F.gelu(self.fc1(x))
      x = F.gelu(self.fc2(x))
      x = F.gelu(self.fc3(x))
      x = F.gelu(self.fc4(x))
      x = F.gelu(self.fc5(x))
      x = F.gelu(self.fc6(x))
      x = F.gelu(self.fc7(x))
      x = self.fc8(x)
      return x

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

model = Net().to(device)
criterion = nn.MSELoss()

wd = 0.01
optimizer = optim.Adam(model.parameters(), lr=0.001, weight_decay=wd)

num_epochs = 2000

# Define the learning rate scheduler
scheduler = optim.lr_scheduler.MultiStepLR(optimizer, milestones=[50, 500, 1000, 1500], gamma=0.2)
# scheduler = optim.lr_scheduler.LambdaLR(optimizer, lr_lambda=lambda epoch: 1 - epoch/(num_epochs-1))


for epoch in range(num_epochs):
    train_loss = 0.0
    model.train()

    for inputs, targets in train_loader:
        inputs, targets = inputs.to(device), targets.to(device)

        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, targets)
        reg_loss = 0.0

        # Adding L2 regularization to the weights of the model
        for param in model.parameters():
            reg_loss += torch.sum(torch.square(param))
        
        loss += wd * reg_loss
        loss.backward()
        optimizer.step()

        train_loss += loss.item()

    train_loss /= len(train_loader)

    with torch.no_grad():
        test_loss = 0.0
        model.eval()

        for inputs, targets in test_loader:
            inputs, targets = inputs.to(device), targets.to(device)

            outputs = model(inputs)
            loss = criterion(outputs, targets)
            test_loss += loss.item()

        test_loss /= len(test_loader)

    scheduler.step()  # Update the learning rate

    if epoch % 25 == 0:
      print(f'Epoch {epoch + 1}, Train Loss: {train_loss:.4f}, Test Loss: {test_loss:.4f}, Learning Rate: {optimizer.param_groups[0]["lr"]:.6f}')

print(f'Epoch {epoch + 1}, Train Loss: {train_loss:.4f}, Test Loss: {test_loss:.4f}, Learning Rate: {optimizer.param_groups[0]["lr"]:.6f}')
```

```python
def predict(sc_height, sc_orient, room_orient, room_width, room_depth):
  scaler.fit(df.iloc[:, :5])
  sample_input = torch.from_numpy(scaler.transform(pd.DataFrame([[sc_height, sc_orient, room_orient, room_width, room_depth]])).astype('float32'))
  sample_input = sample_input.to(device)

  # Make a prediction using the trained model
  model.eval()
  with torch.no_grad():
      predicted_output = model(sample_input)

  # Convert the predicted output tensor to a numpy array and apply inverse transformation
  predicted_output = predicted_output.cpu().numpy()
  scaler.fit(df.iloc[:, 5:])

  predicted_output = scaler.inverse_transform(predicted_output)

  return predicted_output

print(predict(4, 0, -2, 10, 22)[0]) # 2642.697379	2858	2836	1761	1759	799	787	11.716214	11.267217
```
